```python
import pandas as pd
```

```python
file_1 = 'defects_data.csv'
file_2 = 'manufacturing_defect_dataset.csv'
```

```python
df_1 = pd.read_csv(file_1)
df_1.head()
```

```python
df_2 = pd.read_csv(file_2)
df_2.head()
```

```python
print(df_1.shape)
print(df_2.shape)
```

```python
df_2 = df_2.head(1000)
df_2
```

```python
df_1.index.rename('id', inplace=True)
df_2.index.rename('id', inplace=True)
```

```python
main_df = pd.merge(df_1, df_2, on = 'id')
main_df.head()
```

```python
main_df.columns
```

```python
main_df = main_df[[#'defect_id',
       'product_id',
        #'defect_type',
        'defect_date',
       #'defect_location','severity',
       'inspection_method', 'repair_cost',
       'ProductionVolume', 'ProductionCost',
       #'SupplierQuality',
       #'DeliveryDelay',
       'DefectRate', 'QualityScore', 'MaintenanceHours',
       'DowntimePercentage', #'InventoryTurnover', 'StockoutRate',
       #'WorkerProductivity',
       #'SafetyIncidents',
       'EnergyConsumption',
       #'EnergyEfficiency',
       'AdditiveProcessTime', 'AdditiveMaterialCost',
       #'DefectStatus'
                   ]].copy()
```

```python
main_df.info()
```

```python
main_df.head()
```

```python
main_df.columns
```

```python
main_df.describe()
```

```python
main_df = main_df.rename(columns = {'product_id': 'batch_id',
                'defect_type': 'defect_type', 'defect_date':'batch_run_date',
                'inspection_method':'inspection_method',
                'repair_cost':'batch_defects', 'ProductionVolume':'production_volume',
                'ProductionCost':'production_cost', 'DefectRate':'defect_rate',
                'QualityScore':'yield_percentage', 'MaintenanceHours':'maintenance_hours',
                'DowntimePercentage':'down_time_percentage',
                'EnergyConsumption':'energy_consumption',
                'AdditiveProcessTime':'additive_process_time',
                'AdditiveMaterialCost':'additive_material_cost'}).copy()
```

```python
main_df.head()
```

```python
main_df['production_volume']= main_df['production_volume'] *100
main_df['batch_defects']= main_df['batch_defects'] *3.8
 #This will be the number of fails
```

```python
main_df
```

```python
defects = main_df[['inspection_method','batch_defects', 'production_volume']]
defects
```

```python
print(defects.inspection_method.unique())
print(defects.inspection_method.value_counts())
```

```python
vision = defects.where(defects['inspection_method'] == 'Visual Inspection')
sum_vision = vision.batch_defects.sum()
prod_vis = vision.production_volume.sum()

vision = defects.where(defects['inspection_method'] == 'Automated Testing')
sum_auto = vision.batch_defects.sum()
prod_auto = vision.production_volume.sum()

vision = defects.where(defects['inspection_method'] == 'Manual Testing')
sum_man = vision.batch_defects.sum()
prod_man = vision.production_volume.sum()

print(f'{sum_vision}vs{prod_vis}\n{sum_auto}vs{prod_auto}\n{sum_man}vs{prod_man}')

# Yield is simillar accross the line at 97% so could do a cost analysis and
# give reasons as to whi selection is made for CI
# Need a bar chart to show this visual representation
# Calling Vision 1, RFID, Vision 2
```
